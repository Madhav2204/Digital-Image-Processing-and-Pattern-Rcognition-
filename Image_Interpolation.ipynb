# **Image Interpolation**
Python Implementation  of : 

- Nearest Neighbor Interpolation
- Bilinear Interpolation
- Bicubic Interpolation


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
img_url = '/content/gdrive/MyDrive/apple.jpg'
imgg = plt.imread(img_url)
plt.imshow(imgg)
plt.show()
```

Nearest Neighbor Interpolation

```python
near_img = cv2.resize(imgg,None, fx = 10, fy = 10, interpolation = cv2.INTER_NEAREST)
```

```python
plt.imshow(near_img)
plt.show()
```

Clearly, this produces a pixelated or blocky image. Also, it doesn’t introduce any new data.

Bilinear Interpolation

```python
bilinear_img = cv2.resize(imgg,None, fx = 10, fy = 10, interpolation = cv2.INTER_LINEAR)
```

```python
plt.imshow(bilinear_img)
plt.show()
```

This produces a smooth image than the nearest neighbor but the results for sharp transitions like edges are not ideal because the results are a weighted average of 2 surrounding pixels.

Bicubic Interpolation

```python
bicubic_img = cv2.resize(imgg,None, fx = 10, fy = 10, interpolation = cv2.INTER_CUBIC)
```

```python
plt.imshow(bicubic_img)
plt.show()
```

Clearly, this produces a sharper image than the above 2 methods. See the white patch on the left side of the apple. This method balances processing time and output quality fairly well.
